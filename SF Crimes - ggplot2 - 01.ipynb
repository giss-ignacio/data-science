```python
# SF Crimes - ggplot2 - 01.ipynb
# Bubble Plot
# R code
# Ignacio Giss
#

library(ggplot2)
library(RColorBrewer)
# df5 is a sample with the most common crimes in SF by month and year.

ggplot(df5, aes(x = Dates, y = Crimes, label = " ")) +
    geom_point(aes(size = Crimes, colour = Category, alpha=.01)) + 
    geom_text(hjust = 1, size = 3) +
    # size of bubbles from 5 to 20
    scale_size(range = c(5,20)) +
    # size of the legend references for Category 
    guides(colour = guide_legend(override.aes = list(size=5,linetype=0))) +
    # a theme with a white background
    theme_bw() +  
    # colours palette
    scale_color_brewer(palette="Set1") + 
    # x axis labels rotation
    theme(axis.text.x = element_text(angle = 90, hjust = 1))
    
```
