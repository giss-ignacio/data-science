## Clustering Movies

Dividiendo el set de movies en 13 clusters usando K-Means en Matlab.

```python
# finger7.1.ipynb
# Clustering Movies
# Ignacio Giss


# Vamos a usar un set de datos de MovieLens en donde tenemos 
# usuarios, películas y calificaciones. 
# El archivo matrix.zip contiene las filas, columnas y valores de la 
# matriz dispersa de peliculas x usuarios. 
# Deberíamos llegar a una matriz de 3952 filas (películas) x 6040 (usuarios). 
# Queremos agrupar las películas en clusters, tomando a 
# cada película como un punto y sus calificaciones como 
# atributos, se supone que películas con calificaciones similares 
# son semejantes y por eso tiene sentido incluirlas en el mismo cluster. 
from IPython.display import Image
```

```python
'''Imprime las descripciones de las peliculas según su indice'''
def print_movies(movies, res):
    i = 0
    for line in res:
        if i < len(movies):           
            while int(line) > int(movies[i].split(",")[0]):
                i+=1
        print(movies[i])
```
