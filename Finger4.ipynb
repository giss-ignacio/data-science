```python
# Finger4.ipynb
# Ignacio Giss

import matplotlib.pyplot as plot
import matplotlib.cm as cm
import matplotlib.colors
import numpy

%matplotlib inline
```

```python
def getSequences(inWord):
    wordLength = len(inWord)
    return [inWord[0:j+1] for j in xrange(wordLength-1, 0, -1)] 

def countOccurrences(fileIn, bufferSize=100, readBuffer=1000):
    text=fileIn.read().replace('\n', '')
    dataCount = []    
    
    for absPos in range(len(text) - bufferSize - readBuffer):        
        allSeq = getSequences(text[absPos: absPos + bufferSize])
        for seq in allSeq:            
            foundPos = text[absPos + bufferSize : absPos + bufferSize + readBuffer].rfind(seq)            
            if foundPos != -1:                
                dataCount.append((foundPos, len(seq)))
                break
                
    return dataCount
```

```python

def graph(data, alphav=0.1, bufferSize=100, readBuffer=1000):    
    
    plot.figure(figsize=(25,6))
    xData = [x[0] for x in data]    
    yData = [y[1] for y in data]  
    rcol = numpy.arange(len(yData))    
    plot.scatter(xData, yData, edgecolors='none', c=yData, alpha = alphav, s=65)
       
    plot.xlabel('distancia')
    plot.ylabel('longitud')
    plot.colorbar()
    plot.axis([0, readBuffer-40, 0, bufferSize+10])
    
    plot.show()
```

## Shakespeare

```python
# 5.205.583 caracteres, 122.449 lineas
text1 = open("shakespeare.txt")
dataTxt1 = countOccurrences(text1, 120)
graph(dataTxt1, 0.06, 120)
```
