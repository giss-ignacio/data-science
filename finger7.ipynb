## Clustering Movies

Dividiendo el set de movies en 5 clusters usando K-Means en Matlab.

```python
# finger7.ipynb
# Clustering Movies
# Ignacio Giss

from IPython.display import Image
```

```python
'''Imprime las descripciones de las peliculas según su indice'''
def print_movies(movies, res):
    i = 0
    for line in res:
        while int(line) > int(movies[i].split(",")[0]):
            i+=1
        print(movies[i])
```

### Cluster 1

```python
movies_f = open("Movies.dat")
res_f = open("res1.txt")

movies=movies_f.read().split("\n")


res = res_f.read().split()

print_movies(movies, res)
    
Image(filename='c1.png')     
```

### Cluster 2

```python

res_f = open("res2.txt")

res = res_f.read().split()

print_movies(movies, res)
    
Image(filename='c2.png')   
```
